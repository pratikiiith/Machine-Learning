```python
import pandas as pd
import numpy as np
import math
from sklearn.metrics import accuracy_score
import os
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(stop_words="english",decode_error='ignore')

files=[]
content=[]
currlabel={}

directory="/home/pratik/Desktop/2ndSem/Smai-2/Datasets/Question-6/dataset"

for a,b,filename in os.walk(directory):
    files.append(filename)
doc=0
for i in files[0]:
    temp = i.split('_')
    currlabel[doc] = int(temp[1][0])
    doc=doc+1
    newpath = directory + "/" + i
    f = open(newpath,"rb")
    data = f.read()
    content.append(data)
    f.close()
    
X = vectorizer.fit_transform(content).toarray()
print(X.shape)
```

```python
import random
center=[0,0,0,0,0]
j=[]
while(len(j) != 5):
    x = np.random.randint(0,1724)
    if(currlabel[x] not in j):
        center[currlabel[x]-1] = x
        j.append(currlabel[x])

print(center)
```

```python
def score(newlabel):
    ytrue=[]
    ypred=[]
    for i in range(0,len(newlabel)):
        ypred.append(newlabel[i])
        ytrue.append(currlabel[i])
    print(accuracy_score(ytrue, ypred))
    
def calculate(a,b):
    summ=np.square(np.subtract(a,b))
#     summ = np.absolute(summ)
    summ = np.sum(summ)
    summ = math.sqrt(summ)
    return summ

def train(iter):
    newlabel = currlabel.copy()
#     print(newlabel)
    
    newcenter=[]
    for i in center:
        newcenter.append(X[i])
        
    for iteration in range(0,iter):
        for i in range(0,X.shape[0]):
            dist=None
            label=None
            for j in range(0,5):
                temp = calculate(X[i],newcenter[j])
    #             result.append([temp,c])
                if(dist == None):
                    dist = temp
                    label = j+1
                elif(dist > temp):
                    dist = temp
                    label = j+1
            newlabel[i] = label
        count1=0
        count2=0
        count3=0
        count4=0
        count5=0
        vec1=np.zeros(X.shape[1])
        vec2=np.zeros(X.shape[1])
        vec3=np.zeros(X.shape[1])
        vec4=np.zeros(X.shape[1])
        vec5=np.zeros(X.shape[1])

        for i in range(0,X.shape[0]):
            if(newlabel[i] == 1):
                vec1 = np.add(vec1,X[i])
                count1=count1+1
            elif(newlabel[i] == 2):
                vec2 = np.add(vec2,X[i])
                count2=count2+1
            elif(newlabel[i] == 3):
                vec3 = np.add(vec3,X[i])
                count3=count3+1
            elif(newlabel[i] == 4):
                vec4 = np.add(vec4,X[i])
                count4=count4+1
            elif(newlabel[i] == 5):
                vec5 = np.add(vec5,X[i])
                count5=count5+1
        if(count1 != 0):
            vec1 = np.true_divide(vec1, count1)
        else:
            vec1 = newcenter[0]
        if(count2 != 0):
            vec2 = np.true_divide(vec2, count2)
        else:
            vec2 = newcenter[1]
        if(count3 != 0):
            vec3 = np.true_divide(vec3, count3)
        else:
            vec3 = newcenter[2]
        if(count4 != 0):
            vec4 = np.true_divide(vec4, count4)
        else:
            vec4 = newcenter[3]
        if(count5 != 0):
            vec5 = np.true_divide(vec5, count5)
        else:
            vec5 = newcenter[4]
        newcenter=[]
        newcenter.append(vec1)
        newcenter.append(vec2)
        newcenter.append(vec3)
        newcenter.append(vec4)
        newcenter.append(vec5)
#         print(count1,count2,count3,count4,count5)
        score(newlabel)
    return newlabel
```
